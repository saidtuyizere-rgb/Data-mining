```python
#6. classification: with and without behavioural features
behavioral_feature_names = ["raisedhands", "VisITedResources",
                            "AnnouncementsView", "Discussion"]

X_with_bf    = X.copy()
X_without_bf = X.drop(columns=[c for c in behavioral_feature_names if c in X.columns])

scaler = StandardScaler()
X_with_sc    = scaler.fit_transform(X_with_bf)
X_without_sc = StandardScaler().fit_transform(X_without_bf)

cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

def evaluate(clf, X_data, y_data, cv_obj):
    accs = cross_val_score(clf, X_data, y_data, cv=cv_obj, scoring="accuracy")
    prec = cross_val_score(clf, X_data, y_data, cv=cv_obj, scoring="precision_weighted")
    rec  = cross_val_score(clf, X_data, y_data, cv=cv_obj, scoring="recall_weighted")
    f1   = cross_val_score(clf, X_data, y_data, cv=cv_obj, scoring="f1_weighted")
    return {
        "Accuracy":  round(accs.mean() * 100, 1),
        "Precision": round(prec.mean() * 100, 1),
        "Recall":    round(rec.mean()  * 100, 1),
        "F-Measure": round(f1.mean()   * 100, 1),
    }

classifiers = {
    "DT (J48)": DecisionTreeClassifier(random_state=42),
    "ANN":      MLPClassifier(hidden_layer_sizes=(100,), max_iter=500, random_state=42),
    "NB":       GaussianNB(),
}

rows = []
for name, clf in classifiers.items():
    for label, Xd in [("With BF", X_with_sc), ("Without BF", X_without_sc)]:
        res = evaluate(clf, Xd, y, cv)
        for metric, val in res.items():
            rows.append({"Classifier": name, "Features": label,
                         "Metric": metric, "Value": val})

df_results = pd.DataFrame(rows)
```

```python
# Table 3 — Paper replication
pivot_bf  = df_results[df_results["Features"] == "With BF"   ].pivot(
    index="Classifier", columns="Metric", values="Value")
pivot_wbf = df_results[df_results["Features"] == "Without BF"].pivot(
    index="Classifier", columns="Metric", values="Value")

print("Table 3a — With Behavioural Features (BF):")
display(pivot_bf[["Accuracy","Precision","Recall","F-Measure"]])

print("\nTable 3b — Without Behavioural Features (WBF):")
display(pivot_wbf[["Accuracy","Precision","Recall","F-Measure"]])
```

```python
#  Fig 11: BF vs WBF Accuracy Comparison 
clf_names = list(classifiers.keys())
x = np.arange(len(clf_names))
width = 0.35

bf_acc  = [pivot_bf.loc[n,  "Accuracy"] for n in clf_names]
wbf_acc = [pivot_wbf.loc[n, "Accuracy"] for n in clf_names]

fig, ax = plt.subplots(figsize=(8, 5))
bars1 = ax.bar(x - width/2, bf_acc,  width, label="With BF",    color=COLOR_BAR, edgecolor="white")
bars2 = ax.bar(x + width/2, wbf_acc, width, label="Without BF", color="#9FE1CB", edgecolor="white")

for bars in [bars1, bars2]:
    for bar in bars:
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
                f"{bar.get_height():.1f}", ha="center", fontsize=9)

ax.set_xticks(x)
ax.set_xticklabels(clf_names)
ax.set_ylabel("Accuracy (%)")
ax.set_ylim(0, 100)
ax.set_title("Fig 11 · Accuracy: With vs Without Behavioural Features",
             fontweight="bold", pad=12)
ax.legend()
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import cross_validate, StratifiedKFold
from sklearn.preprocessing import LabelEncoder, StandardScaler
import numpy as np
import pandas as pd

def evaluate(clf, X, y, cv):
    scoring = {
        "accuracy":  "accuracy",
        "precision": "precision_weighted",
        "recall":    "recall_weighted",
        "f1":        "f1_weighted",
    }
    scores = cross_validate(clf, X, y, cv=cv, scoring=scoring)
    clf.fit(X, y)
    return {
        "Accuracy":  round(scores["test_accuracy"].mean()  * 100, 2),
        "Precision": round(scores["test_precision"].mean() * 100, 2),
        "Recall":    round(scores["test_recall"].mean()    * 100, 2),
        "F-Measure": round(scores["test_f1"].mean()        * 100, 2),
    }

print("✅ evaluate() defined")
```

```python
# ── Load dataset ──────────────────────────────────────────
df = pd.read_csv("xAPI-Edu-Data.csv")   # ← change if your file has a different name
print("✅ Dataset loaded:", df.shape)
print("   Columns:", df.columns.tolist())

# ── Encode categorical columns ────────────────────────────
le = LabelEncoder()
cat_cols = df.select_dtypes(include="object").columns.tolist()
cat_cols.remove("Class")  # keep target separate

for col in cat_cols:
    df[col] = le.fit_transform(df[col])

# ── Encode target ─────────────────────────────────────────
df["Class"] = df["Class"].map({"H": 2, "M": 1, "L": 0})

# ── Features and target ───────────────────────────────────
X = df.drop(columns=["Class"])
y = df["Class"]

# ── Scale features ────────────────────────────────────────
scaler = StandardScaler()
X_with_sc = scaler.fit_transform(X)

# ── Cross-validation ──────────────────────────────────────
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

print(f"✅ X_with_sc shape : {X_with_sc.shape}")
print(f"✅ y shape         : {y.shape}")
print(f"✅ Classes         : {y.unique()}")
```

```python
import joblib
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier, RandomForestClassifier

ensemble_experiments = {
    "DT (Traditional)":  DecisionTreeClassifier(random_state=42),
    "ANN (Traditional)": MLPClassifier(hidden_layer_sizes=(100,), max_iter=500, random_state=42),
    "NB (Traditional)":  GaussianNB(),
    "DT + Bagging":      BaggingClassifier(DecisionTreeClassifier(random_state=42), n_estimators=10, random_state=42),
    "ANN + Bagging":     BaggingClassifier(MLPClassifier(hidden_layer_sizes=(100,), max_iter=300, random_state=42), n_estimators=10, random_state=42),
    "NB + Bagging":      BaggingClassifier(GaussianNB(), n_estimators=10, random_state=42),
    "DT + Boosting":     AdaBoostClassifier(DecisionTreeClassifier(max_depth=3, random_state=42), n_estimators=50, random_state=42),
    "ANN + Boosting":    AdaBoostClassifier(MLPClassifier(hidden_layer_sizes=(50,), max_iter=200, random_state=42), n_estimators=10, random_state=42),
    "NB + Boosting":     AdaBoostClassifier(GaussianNB(), n_estimators=50, random_state=42),
    "Random Forest":     RandomForestClassifier(n_estimators=100, random_state=42),
}

ens_rows = []
for name, clf in ensemble_experiments.items():
    print(f"  Training: {name} ...")
    res = evaluate(clf, X_with_sc, y, cv)
    res["Classifier"] = name
    ens_rows.append(res)

df_ens = pd.DataFrame(ens_rows).set_index("Classifier")

print("\nTable 4 — Ensemble Results:")
display(df_ens[["Accuracy","Precision","Recall","F-Measure"]])

# ── Save best model ───────────────────────────────────────
best_clf_name = df_ens["Accuracy"].idxmax()
best_clf = ensemble_experiments[best_clf_name]
best_clf.fit(X_with_sc, y)
joblib.dump(best_clf, "student_model.pkl")

print(f"\n✅ Best model  : {best_clf_name}")
print(f"✅ Accuracy    : {df_ens.loc[best_clf_name, 'Accuracy']}%")
print(f"✅ Saved as    : student_model.pkl")
```
